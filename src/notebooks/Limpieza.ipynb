```python
import pandas as pd
import os
#Leer archivo csv
in_dir = "../../data/clean"
# Recuperamos el nombre del archivo en el directorio in_dir
for filename in os.listdir(in_dir):
    if filename.endswith(".csv"):
        in_file = os.path.join(in_dir, filename)
        break
print(in_file)
df = pd.read_csv(in_file, header=0)
df.head()
```

```python
#Eliminamos columna "description"

df_clean = df #Copia de seguridad, para no perder el dataframe original. 

df_clean.drop(columns=["description"], inplace=True)
df_clean.head()
```

```python
#Eliminamos columna "beer name (full)"
df_clean.drop(columns=["beer name (full)"], inplace=True)
df_clean.head()
```
